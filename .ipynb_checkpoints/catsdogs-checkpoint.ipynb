```python
def print_pictures( fnames, complete_dir ):
    nrows = 4
    ncols = 4

    pic_index = 0

    fig = plt.gcf()
    fig.set_size_inches( ncols*4,  nrows*4)

    pic_index += 8

    next_pix = []


    for fname in fnames[ pic_index-8:pic_index ]:
        next_pix.append( complete_dir+'/'+fname)

    for i, img_path in enumerate(next_pix):
        sp = plt.subplot( nrows, ncols,  i + 1 )
        img = mpimg.imread( img_path )
        plt.imshow(img);


plt.show()
        
```

```python
print_pictures(train_cat_fnames, train_cats_dir )
```
